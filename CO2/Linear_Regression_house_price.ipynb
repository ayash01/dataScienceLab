```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../res/houseprice.csv")
df.head()
```

```python
x = df.drop("price", axis = 1)
x.head()
```

```python
labelEncoder = LabelEncoder()

labels = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "parking", "prefarea", "furnishingstatus"]

for label in labels:
    x[label] = labelEncoder.fit_transform(x[label])

x.head()
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)

x[range(0, 3)]
```

```python
y = df["price"]
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 12)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", MSE)
print("Root Mean Squared Error (RMSE):", MSE)
print("R squared (R2):", R2)
```
